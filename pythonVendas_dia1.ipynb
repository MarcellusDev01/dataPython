```python
import pandas as pd
import plotly.express as px

# Links diretos para os arquivos brutos no GitHub
caminhoVendas = "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/Vendas2022.xlsx"
dePara_Local = "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/CadastroLocalidades.xlsx"
cadastroLojas = "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/CadastroLojas.xlsx"

```

```python
workbook = pd.read_excel(caminhoVendas)
display(workbook)
```

```python
workbook_dePara = pd.read_excel(dePara_Local)
display(workbook_dePara)
```

```python
workbookLojas = pd.read_excel(cadastroLojas)
display(workbookLojas)
```

```python
# Mesclar os DataFrames 'workbookLojas' e 'workbook_dePara' usando a coluna 'id Localidade'
cadatroLojasTratado = pd.merge( workbookLojas, workbook_dePara, on='id Localidade', how='inner')
display(cadatroLojasTratado)
```

```python

work_vendasFinal = pd.merge(workbook, cadatroLojasTratado, on='ID Loja', how='inner')
display(work_vendasFinal)
```

```python
# Agrupar os dados de vendas por 'Continente' e somar a quantidade vendida
vendas_por_continente = work_vendasFinal.groupby('Continente')['Qtd Vendida'].sum().reset_index()

display(vendas_por_continente)
```

```python
# Mapear os continentes para um país representativo para a visualização
continente_pais = {
    'América do Norte': 'United States',
    'América do Sul': 'Brazil',
    'Europa': 'France',
    'África': 'Nigeria',
    'Ásia': 'China',
    'Oceania': 'Australia'
}
```

```python
# Adicionar a coluna 'País' ao DataFrame, que mapeia o continente para o país representativo
vendas_por_continente['País'] = vendas_por_continente['Continente'].map(continente_pais)
```

```python
display(vendas_por_continente)
```

```python
# Criar o gráfico de dispersão geográfico com Plotly Express
fig = px.scatter_geo(vendas_por_continente,
                     locations='País',  # Coluna que contém os países para localização
                     locationmode='country names',  # Modo de localização pelos nomes dos países
                     size='Qtd Vendida',  # Tamanho dos círculos proporcional ao volume de vendas
                     hover_name='Continente',  # Nome do continente a ser exibido ao passar o mouse
                     projection='equirectangular',  # Tipo de projeção do mapa (2D plano)
                     title='Volume de Vendas por Continente',  # Título do gráfico
                     size_max=60)  # Tamanho máximo dos círculos

```

```python
# Exibir o gráfico
fig.show()
```
