```python
import pandas as pd
import plotly.express as px

# Links diretos para os arquivos brutos no GitHub
dePara_Local = "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/CadastroLocalidades.xlsx"
cadastroLojas = "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/CadastroLojas.xlsx"
cadastroProdutos = "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/cadastroProdutos.xlsx"
```

```python
arquivoProdutos = pd.read_excel(cadastroProdutos)
display(arquivoProdutos)
```

```python
# Lista de URLs dos arquivos de vendas
arquivos_vendas = [
    "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/Vendas2020.xlsx",
    "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/Vendas2021.xlsx",
    "https://raw.githubusercontent.com/MarcellusDev01/data_Python_vendas/main/Vendas2022.xlsx",
]

# Lista para armazenar os DataFrames
dataframes = []

# Carregar e concatenar todos os arquivos que começam com "Vendas"
for arquivo in arquivos_vendas:
    df = pd.read_excel(arquivo)
    dataframes.append(df)

# Concatenar todos os DataFrames em um único DataFrame
vendas_completas = pd.concat(dataframes, ignore_index=True)

# Exibir o DataFrame final
display(vendas_completas)
```

```python
#Juntando vendas e cadastro  de produtos
vendas_completas_Custo= pd.merge(vendas_completas, arquivoProdutos, on='SKU', how='inner')
display(vendas_completas_Custo)
```

```python
workbook_dePara = pd.read_excel(dePara_Local)
display(workbook_dePara)
```

```python
workbookLojas = pd.read_excel(cadastroLojas)
display(workbookLojas)
```

```python
# Mesclar os DataFrames 'workbookLojas' e 'workbook_dePara' usando a coluna 'id Localidade'
cadastroFinalLojas = pd.merge( workbookLojas, workbook_dePara, on='id Localidade', how='inner')
display(cadastroFinalLojas)
```

```python

work_vendasFinal = pd.merge(vendas_completas_Custo, cadastroFinalLojas, on='ID Loja', how='inner')
display(work_vendasFinal)
work_vendasFinal.to_excel('work_vendasFinal.xlsx', index=False)
```
